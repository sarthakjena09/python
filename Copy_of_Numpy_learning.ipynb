```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    # a. Insert five elements at the beginning
    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # b. Delete an element from the list
    def delete(self, key):
        temp = self.head

        # If head node itself holds the key
        if temp is not None:
            if temp.data == key:
                self.head = temp.next
                temp = None
                return

        # Search for the key to be deleted
        prev = None
        while temp is not None and temp.data != key:
            prev = temp
            temp = temp.next

        # If key was not found
        if temp is None:
            print(f"Element {key} not found in the list.")
            return

        # Unlink the node from linked list
        prev.next = temp.next
        temp = None

    # c. Search for a specific element in the list
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print(f"Element {key} found in the list.")
                return True
            temp = temp.next
        print(f"Element {key} not found in the list.")
        return False

    # d. Print the elements of the list in reverse order
    def print_reverse(self):
        def _reverse_recursive(node):
            if node is None:
                return
            _reverse_recursive(node.next)
            print(node.data, end=" ")

        _reverse_recursive(self.head)
        print()

    # Utility function to print the list
    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

# Testing the implementation
sll = SinglyLinkedList()

# Insert 5 elements
for i in [10, 20, 30, 40, 50]:
    sll.insert_at_beginning(i)

print("Linked List after insertion:")
sll.print_list()

# Delete an element
sll.delete(30)
print("Linked List after deleting 30:")
sll.print_list()

# Search for an element
sll.search(20)
sll.search(100)

# Print in reverse order
print("Linked List in reverse order:")
sll.print_reverse()
```

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    # a. Insert five elements into the tree
    def insert(self, key):
        self.root = self._insert_recursive(self.root, key)

    def _insert_recursive(self, root, key):
        if root is None:
            return Node(key)
        if key < root.key:
            root.left = self._insert_recursive(root.left, key)
        else:
            root.right = self._insert_recursive(root.right, key)
        return root

    # b. Search for a specific element in the tree
    def search(self, key):
        return self._search_recursive(self.root, key)

    def _search_recursive(self, root, key):
        if root is None:
            print(f"Element {key} not found in the tree.")
            return False
        if root.key == key:
            print(f"Element {key} found in the tree.")
            return True
        if key < root.key:
            return self._search_recursive(root.left, key)
        return self._search_recursive(root.right, key)

    # c. Delete an element from the tree
    def delete(self, key):
        self.root = self._delete_recursive(self.root, key)

    def _delete_recursive(self, root, key):
        if root is None:
            return root
        if key < root.key:
            root.left = self._delete_recursive(root.left, key)
        elif key > root.key:
            root.right = self._delete_recursive(root.right, key)
        else:
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left
            temp = self._min_value_node(root.right)
            root.key = temp.key
            root.right = self._delete_recursive(root.right, temp.key)
        return root

    def _min_value_node(self, node):
        current = node
        while current.left is not None:
            current = current.left
        return current

    # d. Print the tree using in-order traversal
    def inorder_traversal(self):
        self._inorder_recursive(self.root)
        print()

    def _inorder_recursive(self, root):
        if root is not None:
            self._inorder_recursive(root.left)
            print(root.key, end=" ")
            self._inorder_recursive(root.right)

    # Find the height of the binary search tree
    def find_height(self):
        return self._find_height_recursive(self.root)

    def _find_height_recursive(self, root):
        if root is None:
            return -1
        left_height = self._find_height_recursive(root.left)
        right_height = self._find_height_recursive(root.right)
        return max(left_height, right_height) + 1

# Testing the implementation
bst = BST()

# Insert 5 elements
for value in [50, 30, 70, 20, 40]:
    bst.insert(value)

print("In-order traversal after insertion:")
bst.inorder_traversal()

# Search for an element
bst.search(40)
bst.search(100)

# Delete an element
bst.delete(30)
print("In-order traversal after deleting 30:")
bst.inorder_traversal()

# Find the height of the tree
print(f"Height of the tree: {bst.find_height()}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
x = np.arange(1, 11)
y = np.random.randint(10, 100, 10)
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 89]

# 1. Line Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Trend')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Line Plot')
plt.legend()
plt.grid(True)
plt.show()

# 2. Histogram
data = np.random.randn(1000)  # Generating random data
plt.figure(figsize=(6, 4))
plt.hist(data, bins=20, color='c', edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.grid(True)
plt.show()

# 3. Bar Chart
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color=['red', 'blue', 'green', 'purple', 'orange'])
plt.xlabel('Categories')
plt.ylabel('Performance')
plt.title('Bar Chart')
plt.grid(axis='y')
plt.show()

# 4. Scatter Plot
x_scatter = np.random.rand(50) * 10
y_scatter = np.random.rand(50) * 10
plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, color='m', marker='x', alpha=0.7)
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Scatter Plot')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample Data
np.random.seed(42)
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C', 'D'], 100),
    'Value': np.random.randn(100) * 10 + 50,
    'Feature1': np.random.randn(100),
    'Feature2': np.random.randn(100),
    'Feature3': np.random.randn(100)
})

# 1. Box Plot
plt.figure(figsize=(6, 4))
sns.boxplot(x='Category', y='Value', data=data, palette='Set2')
plt.title('Box Plot: Distribution across Categories')
plt.grid(True)
plt.show()

# 2. Heatmap
correlation_matrix = data[['Feature1', 'Feature2', 'Feature3']].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap: Correlation between Variables')
plt.show()
```

```python
import numpy as np

# 1. Create a NumPy array and perform basic operations
arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])

print("Array 1:", arr1)
print("Array 2:", arr2)

print("Addition:", arr1 + arr2)
print("Subtraction:", arr1 - arr2)
print("Multiplication:", arr1 * arr2)

# 2. Statistical Measures
data = np.array([12, 15, 14, 10, 18, 20, 25])
print("\nMean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))

# 3. Reshape and Slice
matrix = np.arange(1, 13).reshape(3, 4)
print("\nReshaped Matrix:\n", matrix)
print("Sliced Element (2nd row, 3rd column):", matrix[1, 2])
print("Extracting first two rows:\n", matrix[:2, :])

# 4. Element-wise operations and Broadcasting
arr3 = np.array([1, 2, 3])
print("\nElement-wise square:", arr3 ** 2)
print("Broadcasting (arr3 + 10):", arr3 + 10)

# 5. Mathematical Functions
arr4 = np.array([1, 2, 3, 4])
print("\nExponential:", np.exp(arr4))
print("Logarithm:", np.log(arr4))
```
